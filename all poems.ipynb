```python
import re

def get_last_arabic_char(word):
    # Define a regular expression pattern for Arabic characters
    arabic_pattern = re.compile('[\u0600-\u06FF]+')

    # Find all Arabic characters in the word
    arabic_chars = arabic_pattern.findall(word)

    # Return the last Arabic character if found, otherwise return an empty string
    return arabic_chars[-1] if arabic_chars else ''

def get_last_arabic_chars_in_text(text):
    # Split the text into words
    words = text.split()

    # Get the last Arabic character from each word
    last_arabic_chars = [get_last_arabic_char(word) for word in words]

    return last_arabic_chars

# Example usage
result = get_last_arabic_chars_in_text(text)
print(result)
```

```python
import re

def get_last_arabic_char(word):
    # Define a regular expression pattern for Arabic characters
    arabic_pattern = re.compile('[\u0600-\u06FF]+')

    # Find all Arabic characters in the word
    arabic_chars = arabic_pattern.findall(word)

    # Return the last Arabic character if found, otherwise return an empty string
    return arabic_chars[-1][-1] if arabic_chars else ''

def get_last_arabic_chars_in_text(text):
    # Split the text into words
    words = text.split()

    # Get the last Arabic character from each word
    last_arabic_chars = [get_last_arabic_char(word) for word in words]

    return last_arabic_chars

# Example usage
result = get_last_arabic_chars_in_text(text)
print(result)
```

```python
from collections import Counter
import re

def get_last_arabic_char(word):
    arabic_pattern = re.compile('[\u0600-\u06FF]+')
    arabic_chars = arabic_pattern.findall(word)
    return arabic_chars[-1][-1] if arabic_chars else ''

def get_most_common_last_arabic_char(text):
    words = text.split()
    last_chars = [get_last_arabic_char(word) for word in words]

    # Use Counter to count occurrences of each character
    char_counts = Counter(last_chars)

    # Find the most common character
    most_common_char = char_counts.most_common(1)

    return most_common_char[0][0] if most_common_char else None

# Example usage
result = get_most_common_last_arabic_char(text)
print(result)
```
